# Export Result


```python
def scale(s):
    return int(int(s) * 11.52)

def getBox(data):

    left = scale(data[3])
    top = scale(data[4])
    right = scale(data[5])
    bottom = scale(data[6])
    
    return (left, top, right, bottom)

def createImages(fileName, load_img_dir, save_img_dir):
    f = open(fileName, "r")

    for l in f:
        l = l.replace("\n", "")
        data = l.split(",")
        imgName = data[0].split("/")[-1]
        
        if "add" in imgName or "remove" in imgName:
            continue
        
        imgName = imgName.replace("_small.jpg", "")
        
        print(imgName)
        
        category = int(data[1])
        
        box = getBox(data)
        
        im = Image.open(os.path.join(load_img_dir, imgName + ".jpg"))
        im = im.crop(box)
        
        im.save(os.path.join(save_img_dir, imgName + "_" + '_'.join(data[3:7]) + ".jpg"))
    f.close()
```

# Apply Boxes

```python
def get_boxplots(fileName, load_img_dir, save_img_dir):
    f = open(fileName, "r")
    list_img = []
    my_boxplots = {}
    
    for l in f:
        l = l.replace("\n", "")
        data = l.split(",")
        list_img.append(data[0].split("/")[-1])
    f.close()
    
    
    for i in np.unique(list_img):
        my_boxplots[i] = []
    
    f = open(fileName, "r")
    for l in f:
        l = l.replace("\n", "")
        data = l.split(",")
        imgName = data[0].split("/")[-1]
        
        my_boxplots[imgName].append([float(data[1])]+[int(x) for x in data[3:7]])
    f.close()
    
    for key, values in my_boxplots.items():
        image = imageio.imread(os.path.join(os.path.join(load_img_dir, key)))
    
        bbs_list_prod = []
        bbs_list_label = []
        for row in values:
            if row[0] == 0:
                bbs_list_label.append(ia.BoundingBox(x1=row[1], y1=row[2], x2=row[3], y2=row[4]))
            else:
                bbs_list_prod.append(ia.BoundingBox(x1=row[1], y1=row[2], x2=row[3], y2=row[4]))
        
        bbs_label = ia.BoundingBoxesOnImage(bbs_list_label, shape=image.shape)
        bbs_prod = ia.BoundingBoxesOnImage(bbs_list_prod, shape=image.shape)
        
        image = bbs_label.draw_on_image(image, thickness=2, color=[0, 0, 255])
        image = Image.fromarray(bbs_prod.draw_on_image(image, thickness=2))
        image.save(os.path.join(save_img_dir, key))
    
```
